Given that range ($R = 15\,{\rm m}$), we should be able to say something about the speed with which the blade was launched. Neglecting air resistance, the range $R$ is given by
$$
R = \frac{\upsilon_0^2 \sin \left( 2 \theta \right)}{g},
$$
where $\upsilon_0$ is the magnitude of the initial velocity, $\theta$ is the launch angle (relative to a flat ground), and $g$ is the gravitational acceleration ($3.71\,{\rm m\ s^{-2}}$ for Mars).

Given $R$, we can solve for $\upsilon_0$, assuming a range of launch angles.

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
def v_naught(R, theta, g=3.71):
    # theta in degrees
    # v_0 in units of sqrt(g R)
    return np.sqrt(g*R/np.sin(2.*np.pi/180.*theta))
```

```python
theta = np.arange(10, 80, 1.)
R = 15. # meters

fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111)

ax.plot(theta, v_naught(R, theta), color='k', lw=3, )
ax.grid(True)
ax.set_xlabel(r'$\theta\,\left( {\rm ^\circ} \right)$', fontsize=36)
ax.set_ylabel(r'$\upsilon_0\,\left( {\rm m\ s^{-1}} \right)$', fontsize=36)

ax.tick_params(labelsize=24)
```

$\theta = 45^\circ$ gives the minimum required velocity $\upsilon_0 = 7.5\,{\rm m\ s^{-1}}$.

We should then be able to translate that initial velocity into a rotation speed $\Omega$ (assuming the blade was launched as a result of rotary motion). 
$$
\Omega = \upsilon_0/r,
$$
where $r$ is the distance between the center of mass and rotation axis, which seems to be about $30\,{\rm cm}$ for Ingenuity.

```python
theta = np.arange(10, 80, 1.)
R = 15. # meters
r = 30e-2 #cm to m
Omega_rpm = v_naught(R, theta)/r*(1./2./np.pi)*(60./1.) # rad/s to rpm

fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111)

ax.plot(theta, Omega_rpm, color='k', lw=3, )
ax.grid(True)
ax.set_xlabel(r'$\theta\,\left( {\rm ^\circ} \right)$', fontsize=36)
ax.set_ylabel(r'$\Omega\,\left( {\rm rpm} \right)$', fontsize=36)

ax.tick_params(labelsize=24)
```

which is much smaller than the normal rotation rate of 2400 rpm - https://en.wikipedia.org/wiki/Ingenuity_(helicopter).

What about including air drag? Actually looks like air drag is enormous for such a small blade:
$$
F_{\rm drag} = c upsilon^2 \approx \rho A \upsilon^2 = \rho A \Omega^2 r^2 = \left( 0.02\,{\rm kg\ m^{-3}} \right) \left( 8.4\,{\rm cm} \right) \left( 60\,{\rm cm} \right) \left( 250\,{\rm rad\ s^{-1}} \right)^2 \left( 30\,{\rm cm} \right)^2 = 5.6\,{\rm N},
$$

as compared to a weight of 0.15 N.
